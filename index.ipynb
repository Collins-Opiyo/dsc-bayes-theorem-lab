## Define a custom function for Bayes' theorem

To start, write a function, `bayes()`, which takes in the probability of A, the probability of B, and the probability of B given A. From this, the function should then return the conditional probability of A, given that B is true.

```python
def bayes(P_a, P_b, P_b_given_a):
    # Your code here
    # Calculate P(A | B) using Bayes' theorem
    P_a_given_b = (P_b_given_a * P_a) / P_b
    return P_a_given_b
```

## Skin Cancer

After a physical exam, a doctor observes a blemish on a client's arm. The doctor is concerned that the blemish could be cancerous, but tells the patient to be calm and that it's probably benign. Of those with skin cancer, 100% have such blemishes. However, 20% of those without skin cancer also have such blemishes. If 15% of the population has skin cancer, what's the probability that this patient has skin cancer? 

> Hint: Be sure to calculate the overall rate of blemishes across the entire population.

```python
def calculate_probability_cancer_given_blemish():
    # Given probabilities
    P_cancer = 0.15
    P_not_cancer = 1 - P_cancer
    P_blemish_given_cancer = 1.0
    P_blemish_given_not_cancer = 0.20
    
    # Calculate overall probability of having a blemish
    P_blemish = P_blemish_given_cancer * P_cancer + P_blemish_given_not_cancer * P_not_cancer
    
    # Apply Bayes' theorem
    P_cancer_given_blemish = (P_blemish_given_cancer * P_cancer) / P_blemish
    
    return P_cancer_given_blemish

# Calculate the probability
result = calculate_probability_cancer_given_blemish()
print(f"Probability of having skin cancer given a blemish: {result}")                                   
```

## A diagnostic test

A diagnostic test is advertised as being 99% accurate 

* If a patient has the disease, they  will test positive 99% of the time 

* If they don't have the disease, they will test negative 99% of the time  

* 1% of all people have this disease 

If a patient tests positive, what is the probability that they actually have the disease?

```python
# Your solution P(Disease | positive test)
def calculate_probability_disease_given_positive_test():
    # Given probabilities
    P_disease = 0.01  # Probability of having the disease
    P_positive_given_disease = 0.99  # Probability of testing positive given disease
    P_negative_given_not_disease = 0.99  # Probability of testing negative given not disease
    
    # Calculate overall probability of testing positive
    P_positive_test = (P_positive_given_disease * P_disease +
                       (1 - P_negative_given_not_disease) * (1 - P_disease))
    
    # Apply Bayes' theorem
    P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive_test
    
    return P_disease_given_positive

# Calculate the probability
result = calculate_probability_disease_given_positive_test()
print(f"Probability of having the disease given a positive test: {result}")
```
